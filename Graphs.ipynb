## Plot God mentions graph

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# To force integers on graph
from matplotlib.figure import Figure
from matplotlib.ticker import MaxNLocator
```

```python
god = pd.read_csv("./god_mentions.csv")
god
```

```python
colour1 = "#E24220"
colour2 = "#3C97C5"
colour3 = "#48414A"
```

```python
# To make the graph bigger (USE IT BEFORE plt.plot)
plt.figure(figsize=(13,8))

plt.plot(god.YEAR, god.FRANCE, colour1, label='France')
plt.plot(god.YEAR, god.ITALY, colour2, label='Italy')
plt.plot(god.YEAR, god.DENMARK, 'm-', label='Denmark')
plt.plot(god.YEAR, god.NORWAY, colour3, label='Norway')
plt.plot(god.YEAR, god.SPAIN, 'g-', label='Spain')
plt.plot(god.YEAR, god.UK, 'y-', label='United Kingdom') # "Commonwealth"

plt.title('Frequency of mention of the word "God"', fontsize=15)

fg = Figure()

# To name the axes
plt.xlabel('Year')
plt.ylabel('Frequency of mention of the word "God"')

# To create a legend with the labels
plt.legend()

plt.xticks(god.YEAR[::5].tolist()+[2020]) 

plt.savefig('./god_mentions.PNG', dpi=300)

plt.show()
```

## God mentions subplots

```python
colour_f = "#B20A1E"
colour_i = "#292A83"

colour_d = "#3F6265"
colour_uk = "#A39911"
```

```python
# figure size = (width,height)
f1 = plt.figure(figsize=(33,13))
# f2 = plt.figure(figsize=(30,13))

# plot one
# Plot column labeled TIME from csv file and color it red
# subplot(2 Rows, 3 Columns, First subplot,)
ax1 = f1.add_subplot(2,3,1)
ax1.plot(god['YEAR'], god['FRANCE'], label = 'France', color = colour_f, markevery = 10)  

# Plot one options
ax1.legend(loc='upper right', fontsize=20)  #'large'
#  ax1.set_title('France')
ax1.set_xlabel('Year', fontsize= 20)
ax1.set_ylabel('Number of occurrences', fontsize= 20)
# ax1.set_xlim([0,200])
ax1.set_ylim([-0.5,10])   

# plot two
# plot column labeled TIME from csv file and color it green
# subplot(2 Rows, 3 Columns, Second subplot)
ax2 = f1.add_subplot(2,3,2)
ax2.plot(god['YEAR'], god['ITALY'], label = 'Italy', color = colour_i, markevery = 10)

# Plot two options
ax2.legend(loc='upper right', fontsize=20)   
# ax2.set_title('Title for plot two ')
ax2.set_xlabel('Year', fontsize= 20)
ax2.set_ylabel('Number of occurrences', fontsize= 20)
# ax2.set_xlim([0,200])
ax2.set_ylim([-0.5,10])   

# plot three
# plot column labeled TIME from csv file and color it blue
# subplot(2 Rows, 3 Columns, Third subplot)
ax3 = f1.add_subplot(2,3,3)
ax3.plot(god['YEAR'], god['DENMARK'], label = 'Denmark', color = colour_d, markevery = 10)  

# Plot three options
ax3.legend(loc='upper right', fontsize=20)
# ax3.set_title('Title for plot three ')
ax3.set_xlabel('Year', fontsize= 20)
ax3.set_ylabel('Number of occurrences', fontsize= 20)
ax3.set_ylim([-0.5,10]) 

# plot four
# plot column labeled TIME from csv file and color it purple
# subplot(2 Rows, 3 Columns, Fourth subplot)
ax4 = f1.add_subplot(2,3,4)
ax4.plot(god['YEAR'], god['NORWAY'], label = 'Norway', color = "m")

# Plot four options
ax4.legend(loc='upper right', fontsize=20)
# ax3.set_title('Title for plot three ')
ax4.set_xlabel('Year', fontsize= 20)
ax4.set_ylabel('Number of occurrences', fontsize= 20)
ax4.set_ylim([-0.5,10]) 

# plot five
# plot column labeled TIME from csv file and color it cyan
# subplot(2 Rows, 3 Columns, Fifth subplot)
ax5 = f1.add_subplot(2,3,5)
ax5.plot(god['YEAR'], god['SPAIN'], label = 'Spain', color = "g")

# Plot five options
ax5.legend(loc='upper right', fontsize=20)
# ax3.set_title('Title for plot three ')
ax5.set_xlabel('Year', fontsize= 20)
ax5.set_ylabel('Number of occurrences', fontsize= 20)
ax5.set_ylim([-0.5,10]) 

# plot six
# plot column labeled TIME from csv file and color it black
# subplot(2 Rows, 3 Columns, Sixth subplot)
ax6 = f1.add_subplot(2,3,6)
ax6.plot(god['YEAR'], god['UK'], label = 'United Kingdom', color = colour_uk)    # * because of "Commonwealth"

# Plot six options
ax6.legend(loc='upper right', fontsize=20)
# ax3.set_title('Title for plot three ')
ax6.set_xlabel('Year', fontsize= 20)
ax6.set_ylabel('Number of occurrences', fontsize= 20)
ax6.set_ylim([-0.5,10]) 

# general title
plt.suptitle('Frequency of mention of the word "God"', fontsize=28, color='black') 

# Save figure
plt.savefig('./graphs/god_mentions_dark.PNG', dpi=300)
```

## Plot the positive emotion changes

```python
posit = pd.read_csv("./LIWC/LIWC_comparison.csv")
posit
```

```python
# To make the graph bigger (USE IT BEFORE plt.plot)
plt.figure(figsize=(13,8))

plt.plot(posit.YEAR, posit.FRA_pos, colour1, label='France')
plt.plot(posit.YEAR, posit.ITA_pos, colour2, label='Italy')
plt.plot(posit.YEAR, posit.SPA_pos, 'g-', label='Spain')
plt.plot(posit.YEAR, posit.UK_pos, 'y-', label='United Kingdom') 

plt.title('Change of positive emotion rate over time', fontsize=15)

fg = Figure()

# To name the axes
plt.xlabel('Year', fontsize= 15)
plt.ylabel('Positive emotion rate', fontsize= 15)

# To create a legend with the labels
plt.legend()

plt.yticks(range(9))

plt.xticks(posit.YEAR[::5].tolist()+[2020]) 

plt.savefig('./graphs/positive_emotion.PNG', dpi=300)

plt.show()
```
